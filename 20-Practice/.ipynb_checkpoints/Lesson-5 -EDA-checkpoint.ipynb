# Univariate

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diamonds.csv')
```

```python
df.head()
```

```python
# Countplot
```

```python
sns.countplot(x='cut',data=df)
```

```python
df['cut'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
#histogram
```

```python
plt.hist(df['depth'])
```

```python
sns.distplot(df['depth'])
```

```python
#boxplot
```

```python
sns.boxplot(df['depth'])
```

```python
sns.boxplot(df['table'])
```

# Mutivariate

```python
#scatterplot
```

```python
sns.scatterplot(data=df,x='depth',y='price',hue='color',style='depth',size='x')
```

```python
#barplot
```

```python
sns.barplot(data=df,x='cut',y='price',hue='color')
```

```python
#boxplot -numerical-categorical
```

```python
sns.boxplot(data=df,x='cut',y='price',hue='color')
```

```python
sns.pairplot(data=df,hue = 'color')
```
