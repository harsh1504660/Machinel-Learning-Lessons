```python
import pandas as pd
import requests as req
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/Cricket'
html_content = req.get(url).text
```

```python
soup = BeautifulSoup(html_content,'lxml')
```

```python
print(soup.prettify)
```

```python
for i in soup.find_all(class_='infobox vcard'):
    print(i.text.strip())
```
