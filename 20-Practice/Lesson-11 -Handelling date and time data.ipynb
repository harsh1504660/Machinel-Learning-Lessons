# Date and time

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Insurance_Company_Complaints__Resolutions__Status__and_Recoveries.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Opened']=df['Opened'].astype('datetime64[ns]')
```

```python
df.info()
```

```python
# 1 extract year
```

```python
df.drop(columns=['Coverage','SubCoverage','Reason','SubReason','Disposition','Company','Recovery'],inplace=True)
```

```python
df['opened_year'] = df['Opened'].dt.year
```

```python
df.head()
```

```python
# 2 extract month
df['opened_month'] = df['Opened'].dt.month
```

```python
df.head()
```

```python
# extract days
df['opened_day'] = df['Opened'].dt.day
```

```python
df.head()
```

```python
# name of day
df['opened_day_name'] = df['Opened'].dt.day_name()
```

```python
df.head()
```

```python
# day of week
df['opened_dayofweek'] = df['Opened'].dt.dayofweek
df.head()
```
