```python
from sklearn.preprocessing import FunctionTransformer
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('diamonds.csv')
```

```python
df.head()
```

```python
trf = FunctionTransformer(func=np.log,)
```

```python
df = df.replace(0,15)
```

```python
new_df = trf.fit_transform(df.drop(columns=['cut','clarity','color']))
```
