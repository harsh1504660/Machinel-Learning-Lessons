# Working with dates

```python
# converting to date time datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

### 1.Exreact year

```python
date['date_year'] = date['date'].dt.year

date.head()
```

### 2.Extract month

```python
date['date_month'] = date['date'].dt.month

date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

### 3.Extract days

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
# day of week
date['date_day_of_week'] = date['date'].dt.dayofweek
date.head()
```

```python
# day of week name
date['date_name_of_day'] = date['date'].dt.day_name()
date.head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_name_of_day'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

### 4.Week of the year

```python

date['date_week'] = date['date'].dt.week

date.drop(columns=['product_id','city_id','orders']).head()
```

### 5.Extract quarter of the year

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

### 6.Extract semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

### 7.Extract time elapsed between days

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed

np.round((today -date['date']) / np.timedelta64(1, 'M'),0)
```
