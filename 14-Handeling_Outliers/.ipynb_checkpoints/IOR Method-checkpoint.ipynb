```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(16,5))
plt.subplots(nrows=1, ncols=1)
sns.distplot(df['cgpa'])

plt.subplots()
sns.distplot(df['placement_exam_marks'])
plt.show()
```

Placement_exam_mark is skewed which means we can solv it by iqr methode

```python
df['placement_exam_marks'].describe()
```

```python
sns.boxplot(df['placement_exam_marks'])
```

```python
#Finding the iqr
percentile25=df['placement_exam_marks'].quantile(0.25)
percentile75=df['placement_exam_marks'].quantile(0.75)
```

```python
percentile25,percentile75
```

```python
# finding iqr by using describe function itself
percentile25 = 17.0
percentile75=44.0
```

```python
iqr = percentile75-percentile25
```

```python
iqr
```

```python
upper_limit = percentile75 + (1.5*iqr)
lower_limit = percentile25 - (1.5*iqr)
```

```python
print('lower limit is ',lower_limit)
print('upper limit is ', upper_limit)
```

# finding outliers

```python
df[df['placement_exam_marks'] > upper_limit]
```

```python
df[df['placement_exam_marks'] < lower_limit]
```

# Triming

```python
new_df = df[df['placement_exam_marks'] < upper_limit]
```

```python
new_df.shape
```

```python
# Comparing

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['placement_exam_marks'])

plt.subplot(2,2,2)
sns.boxplot(df['placement_exam_marks'])

plt.subplot(2,2,3)
sns.distplot(new_df['placement_exam_marks'])

plt.subplot(2,2,4)
sns.boxplot(new_df['placement_exam_marks'])

plt.show()
```

# capping 

```python
new_df_cap = df.copy()

for i in range(1,985):
    if df['placement_exam_marks'][i] > upper_limit:
        new_df_cap['placement_exam_marks'][i] = upper_limit
    else :
        if df['placement_exam_marks'][i] <lower_limit:
            new_df_cap['placement_exam_marks'][i] = lower_limit
        else :
            new_df_cap['placement_exam_marks'][i]=df['placement_exam_marks'][i]
```

```python
new_df_cap.describe()
```

```python
# Comparing

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['placement_exam_marks'])

plt.subplot(2,2,2)
sns.boxplot(df['placement_exam_marks'])

plt.subplot(2,2,3)
sns.distplot(new_df_cap['placement_exam_marks'])

plt.subplot(2,2,4)
sns.boxplot(new_df_cap['placement_exam_marks'])

plt.show()
```

```python
new_df_cap.shape
```

```python
# methode 2 
new_df_cap = df.copy()

new_df_cap['placement_exam_marks'] = np.where(
    new_df_cap['placement_exam_marks'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap['placement_exam_marks'] < lower_limit,
        lower_limit,
        new_df_cap['placement_exam_marks']
    )
)
```
