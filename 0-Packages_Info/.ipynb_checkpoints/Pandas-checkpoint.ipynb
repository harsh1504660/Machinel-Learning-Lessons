```python
vk
```

```python
vk_bteam = vk.groupby('bowling_team')
```

```python
vk_bteam.sum()['batsman_runs'].sort_values(ascending=False).head(3)
```

```python
#sol:
vk = deli[deli['batsman']=='V Kohli']
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
def topruns(player):
    pl = deli[deli['batsman']==player]
    op=  pl.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(3)
    return print(op)
```

```python
topruns('CH Gayle')
```

## Merge

```python
new = deli.merge(df,left_on='match_id',right_on='id')
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season').reset_index().drop('index',axis=1)
```

# Pivot Table

```python
mask = deli['batsman_runs']==6
six = deli[mask]
six.head()
```

```python
pt = six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')
pt
```

```python
import seaborn as sns
```

```python
sns.heatmap(pt)
```
