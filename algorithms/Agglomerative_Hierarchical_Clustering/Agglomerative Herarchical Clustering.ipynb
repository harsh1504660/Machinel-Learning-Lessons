```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
data = df.iloc[:,3:5].values
```

```python
data
```

```python
import scipy.cluster.hierarchy as shc
```

```python
y=[200,200,200,200,200,200]
x=[50,300,500,600,1000,1900]
plt.figure(figsize=(10,7))
plt.title('dendogram')
dend = shc.dendrogram(shc.linkage(data,method='ward'))
plt.plot(x,y,color='red')
plt.annotate("cut", xy=(1010,200), xytext=(1200,250),
            arrowprops=dict(arrowstyle="->"))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
labels_=cluster.fit_predict(data)
```

```python
labels_
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')
```
